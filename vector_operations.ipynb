## Matrix Theory: Vector Operations Visualization

This notebook demonstrates basic vector operations such as vector
addition, subtraction, and scalar multiplication using Python and
Matplotlib.

**Acknowledgement:**  
The base code used in this notebook was provided by the course instructor
as part of the Matrix Theory curriculum.

I have executed the code, added comments, and made minor edits for better
clarity and understanding.


```python
import numpy as np
import matplotlib.pyplot as plt
#define vectors
v1=np.array([3,2])
v2=np.array([1,4])
k=2
#vector operations
v_add=v1+v2
v_sub=v1-v2
v_scalar=k*v1
#create figure
fig,ax=plt.subplots(2,2,figsize=(10,8))
#define limits of axes
def setup_axis(a,title):
    a.set_xlim(-8,8)
    a.set_ylim(-8,8)
    a.set_aspect('equal',adjustable='box')
    a.grid(True)
    a.set_title(title)
#plot1
ax[0,0].quiver(0,0,v1[0],v1[1],
                color='blue',
                angles='xy',
                scale_units='xy',
                scale=1,
                label='v1')
ax[0, 0].quiver(0, 0, v2[0], v2[1],
                color='green',
                angles='xy',
                scale_units='xy',
                scale=1,
                label='v2')
setup_axis(ax[0,0],"original vectors")
ax[0,0].legend(loc='upper left',fontsize=9)
#plot2
ax[0, 1].quiver(0, 0, v1[0], v1[1],
                color='blue',
                angles='xy',
                scale_units='xy',
                scale=1,
                label='v1')
ax[0, 1].quiver(0, 0, v2[0], v2[1],
                color='green',
                angles='xy',
                scale_units='xy',
                scale=1,
                label='v2')
ax[0, 1].quiver(0, 0, v_add[0], v_add[1],
                color='red',
                angles='xy',
                scale_units='xy',
                scale=1,
                label='v1 + v2')
setup_axis(ax[0, 1], "Vector Addition")
ax[0, 1].legend(loc='upper left', fontsize=9)
#plot3
ax[1, 0].quiver(0, 0, v1[0], v1[1],
                color='blue',
                angles='xy',
                scale_units='xy',
                scale=1,
                label='v1')
ax[1,0].quiver(0,0,v2[0],v2[1],
               color='green',
               angles='xy',
               scale_units='xy',
               scale=1,
               label='v2')
ax[1,0].quiver(0,0,-v2[0],-v2[1],
               color='black',
               angles='xy',
               scale_units='xy',
               scale=1,
               label='-v2')
ax[1, 0].quiver(0, 0, v_sub[0], v_sub[1],
                color='purple',
                angles='xy',
                scale_units='xy',
                scale=1,
                label='v1-v2')
setup_axis(ax[1, 0], "Vector Subtraction")
ax[1, 0].legend(loc='upper left', fontsize=9)
#plot4
ax[1, 1].quiver(0, 0, v1[0], v1[1],
                color='blue',
                angles='xy',
                scale_units='xy',
                scale=1,
                label='v1')
ax[1, 1].quiver(0, 0, v_scalar[0], v_scalar[1],
                color='orange',
                angles='xy',
                scale_units='xy',
                scale=1,
                label='2 x v1')
setup_axis(ax[1, 1], "Scalar Multiplication")
ax[1, 1].legend(loc='upper left', fontsize=9)
plt.tight_layout()
plt.show()
```
